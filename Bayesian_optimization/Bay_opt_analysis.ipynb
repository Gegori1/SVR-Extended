```python
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt 
import cufflinks as cf
cf.go_offline()
import plotly.express as px
```

# Boston

## load data

```python
jsn_path = Path().resolve() / "Logs" / "Boston_leo.json"
data = pd.read_json(jsn_path , lines = True).drop(["datetime"], axis = 1)
```

## convert data

```python
params = pd.json_normalize(data['params'])
```

```python
df = pd.concat([data[["target"]]*-1, params], axis = 1)
```

## plot

```python
df1 = df.query("target > 0.8")
df2 = df.query("target > 0.7")
```

```python
px.scatter_3d(df1, x = "lamda", y = "gamma", z = "epsilon", color = "target", hover_data = ["C"])
```

```python
px.scatter(df, x="gamma", y="epsilon", color="target")
```
