```python
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt 
import plotly.express as px
```

# $R^2$ Metric

## Boston without symbolic transformation

### load data

```python
jsn_path = Path().resolve() / "Logs" / "Boston_r2_leo.json"
data = pd.read_json(jsn_path , lines = True).drop(["datetime"], axis = 1)
```

### convert data

```python
params = pd.json_normalize(data['params'])
```

```python
df = pd.concat([data[["target"]], params], axis = 1)
```

### plot

```python
df1 = df.query("target > 0.8")
df2 = df.query("target > 0.7")
```

```python
px.scatter_3d(df1, x = "lamda", y = "gamma", z = "epsilon", color = "target", hover_data = ["C"])
```

```python
px.scatter(df, x="gamma", y="epsilon", color="target")
```

```python
data = df.query("target > 0 & target <= 1")
```

```python
data.sort_values("target", ascending = False).head(2)
```

# MAE

## Boston without symbolic transformation

### load data

```python
jsn_path = Path().resolve() / "Logs" / "Boston_mae_leo.json"
data = pd.read_json(jsn_path , lines = True).drop(["datetime"], axis = 1)
```

### convert data

```python
params = pd.json_normalize(data['params'])
```

```python
df = pd.concat([data[["target"]]*-1, params], axis = 1)
df[["max_depth", "n_estimators"]] = df[["max_depth", "n_estimators"]].astype("int")
```

### plot

```python
df1 = df.query("target < 2.6")
df2 = df.query("target < 2.2")
```

```python
px.scatter_3d(df1, x = "n_estimators", y = "max_depth", z = "target", color = "target")
```

```python
px.scatter(df1, x="max_depth", y="n_estimators", color="target")
```

```python
df.sort_values("target", ascending = True).head(20).agg(("max", "min"))
```
