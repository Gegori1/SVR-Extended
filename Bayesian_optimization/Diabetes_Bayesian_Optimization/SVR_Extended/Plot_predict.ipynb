```python
import pandas as pd
import numpy as np
import json
import plotly.express as px
from pathlib import Path
from sklearn.linear_model import Ridge
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
import plotly.graph_objects as go
```

### load data

```python
jsn_path = Path().resolve() / "Logs" / "Boston_r2_let.json"
data = pd.read_json(jsn_path , lines = True).drop(["datetime"], axis = 1)
```

### convert data

```python
params = pd.json_normalize(data['params'])
```

```python
df = pd.concat([data[["target"]], params], axis = 1).query("target > 0")
df.target = df.target*-1
```

```python
X_train = df.drop("target", axis = 1)
y_train = df.target
```

```python
Ce = np.linspace(40, 100, 100)
gam = np.linspace(0.056, 0.1, 100)
eps = np.linspace(5, 25, 100)
# lam = np.linspace(5e-5, 0.4, 100)
Eps, ga = np.meshgrid(eps, gam)
eps, gam = np.ravel(Eps), np.ravel(ga)
lam = [40e-5]#*gam.size
# Ce = [82]#*gam.size
# eps = [25]*Ce.size
```

```python
X_test = pd.DataFrame({"lamda" : lam, "C" : Ce, "gamma" : gam, "epsilon" : eps})
```

```python
model = CatBoostRegressor().fit(X_train, y_train)
y_test = model.predict(X_test)
```

```python
Z = y_test.reshape(Eps.shape)
```

```python
fig = go.Figure(
    data=[go.Surface(
        z=Z, 
        x=Eps, 
        y=ga, 
        hovertemplate = 'C: %{x:.2f}<br>gamma: %{y:.5f}<br>target: %{z:.5f}<extra></extra>',
        opacity = 1
        )
    ]
)
fig.update_layout(scene = dict(xaxis_title = "C", yaxis_title = "gamma", zaxis_title = "target"))
fig.show()
```
